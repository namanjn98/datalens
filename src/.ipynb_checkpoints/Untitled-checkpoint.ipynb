```python
annotations = json.load(open("/Users/namanjain/Downloads/annotations/instances_val2017.json", 'r'))
```

```python
image_id_map = {}
for i in annotations['images']:
    image_id_map[i['id']] = i['file_name']
```

```python
len(image_id_map)
```

```python
image_wise_annotations = {}
for i in annotations['annotations']:
    
    annot = {'bbox': i['bbox'], 'category_id': i['category_id']}
    try:
        filename = image_id_map[i['image_id']]
    except:
        print(i['id'])
        continue
    try:
        image_wise_annotations[filename].append(annot)
    except:
        image_wise_annotations[filename] = annot
```

```python
len(image_wise_annotations)
```
